# Report for Exploratory Data Analysis on Building Permits

The deductions below were made, after performing an Exploratory Data Analysis on the 
Building Permits dataset, by data visualization:

1. The Permit Type 8 has the highest count of permits filed.

2. The Financial District/South Beach Neighborhood has the highest building permits filed.

3. About 90,000 of the permits filed have their current status as complete, while around 80,000 were just issued.

4. The count of fire related permits filed is about 10% of the whole dataset count. This means about 1 in every 10 buildings which filed for permits may have had fire related hazards, which could have prompted taking safety precautions.

5. The Permit Type 8 has the highest cost estimate.

6. Building that have no plans to add more units have highest cost estimates.

7. The highest cost estimates are for buildings with plans to add more stories.

8. Pacific Heights is the Neigborhood with the highest cost estimate, or have the most construction to be done.

9. There are very minimal correlations between the categories in the dataset, and so may not be able to influence each other evidently.
